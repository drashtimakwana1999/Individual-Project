<b>
Q1 The Movie Database is supported by volunteers going out into the world, collecting data, and
entering it into the database. This includes financial data, such as movie budget and revenue. If you
wanted to know which movies are still missing data, you could use a left join to identify them.
Practice using a left join by merging the movies table and the financials table. Count the missing rows with left join.</b>

```python
import pandas as pd

# Load data
movies_df = pd.read_csv(r"C:/Users/Adarsh/Desktop/Drashti/Data Analysis tools analytics/movies.csv")
financials_df = pd.read_csv(r"C:/Users/Adarsh/Desktop/Drashti/Data Analysis tools analytics/financials.csv")

# Merge movies with financials using a left join on 'id'
merged_df = movies_df.merge(financials_df, on='id', how='left')

# Count missing financial details (NaN values in the 'budget' column)
missing_count = merged_df['budget'].isna().sum()

# Print result
print(f"Total movies missing financial details: {missing_count}")
```

<h1>Question #2:</h1>
<b>
Setting how=‘left’ with the .merge()method is a useful technique for enriching or enhancing a
dataset with additional information from a different table. In this exercise, you will start off with a
sample of movie data from the movie series Toy Story. Your goal is to enrich this data by adding
the marketing tag line for each movie. You have to compare the results of a left join versus an inner
join.</b>
</div>


```python
import pandas as pd

# Load data
movies_df = pd.read_csv(r"C:/Users/Adarsh/Desktop/Drashti/Data Analysis tools analytics/movies.csv")
taglines_df = pd.read_csv(r"C:/Users/Adarsh/Desktop/Drashti/Data Analysis tools analytics/taglines.csv")

# Perform a left join on 'id'
left_merged_df = movies_df.merge(taglines_df, on="id", how="left")

# Perform an inner join on 'id'
inner_merged_df = movies_df.merge(taglines_df, on="id", how="inner")

# Display the number of rows in each join
left_join_count = len(left_merged_df)
inner_join_count = len(inner_merged_df)

print(f"Number of rows after the left join: {left_join_count}")
print(f"Number of rows after the inner join: {inner_join_count}")
```

<h1>Question #3:</h1>
<b>
 Find which movies are considered only science fiction movies.
Once you have this table, you can merge the movies table in to see the movie names.</b>
</div>

```python
import pandas as pd

# Load the dataset
genres_df = pd.read_csv(r"C:/Users/Adarsh/Desktop/Drashti/Data Analysis tools analytics/movie_to_genres.csv")
movies_df = pd.read_csv(r"C:/Users/Adarsh/Desktop/Drashti/Data Analysis tools analytics/movies.csv")

# Evaluate how many different genres each film has
genre_counts_df = genres_df.groupby("movie_id")["genre"].nunique()

# Filter movies that have only one genre, specifically "Science Fiction"
sci_fi_movies_df = genres_df[genres_df["genre"] == "Science Fiction"]
sci_fi_movies_df = sci_fi_movies_df[sci_fi_movies_df["movie_id"].isin(genre_counts_df[genre_counts_df == 1].index)]

# Merge with the movies dataset to get movie names
sci_fi_movies_df = sci_fi_movies_df.merge(movies_df, left_on="movie_id", right_on="id", how="left")

# Select relevant columns
sci_fi_movies_df = sci_fi_movies_df[["id", "title", "release_date", "popularity"]]

# Display the first few results
print(sci_fi_movies_df.head())
```

<h1>Question #4:</h1>
<b>
 An outer join returns all rows from both merged tables and null where they do not match, you can use it to find rows that do not have a match in the other table. To try for yourself, you have been given two tables with a list of actors from two popular movies: Iron Man 1 and Iron Man 2. Most of the actors played in both movies. Use an outer join to
find actors who did not act in both movies.</b>
</div>


```python
import pandas as pd

# Load the datasets
iron1_actors_df = pd.read_csv(r"C:/Users/Adarsh/Desktop/Drashti/Data Analysis tools analytics/iron_1_actors.csv")
iron2_actors_df = pd.read_csv(r"C:/Users/Adarsh/Desktop/Drashti/Data Analysis tools analytics/iron_2_actors.csv")

# Perform an outer join on 'id' to find actors who are not in both films
merged_actors_df = iron1_actors_df.merge(iron2_actors_df, on="id", how="outer", suffixes=("_iron1", "_iron2"))

# Identify actors who appeared in only one of the two movies (i.e., have NaN in one column)
unique_actors_df = merged_actors_df[merged_actors_df.isna().any(axis=1)]

# Select relevant columns
unique_actors_df = unique_actors_df[["id", "name_iron1", "name_iron2"]]

# Display the first few results
print(unique_actors_df.head())
```
